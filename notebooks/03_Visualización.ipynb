### 1. Carga de datos ###

```python
import pandas as pd
import seaborn as sns
df = pd.read_csv('../data/processed/Student_performance_data.csv')
```

### 2. Graficos de GPA vs variables categoricas

#### 1. GPA vs Genero

```python
sns.catplot(data=df, x="Gender", y="GPA", hue="Gender", kind="violin");
```

```python
sns.catplot(data=df, x="Ethnicity", y="GPA", hue="Ethnicity", kind="box");
```
