### 1. Carga de datos ###

Se importan las librerías y se carga dataset limpio.

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib as plt
df = pd.read_csv('../data/processed/Student_performance_data.csv')
```

Se establece tema y paleta de colores para los gráficos

```python
sns.set_theme(style="whitegrid")
sns.set_palette("hls", n_colors=None, desat=None, color_codes=False)
sns.color_palette("hls")
```

Orden para la leyenda de GradeClass

```python
gradeClass = ["A", "B", "C", "D", "E", "F"]
```

### 2. Análisis de distribución de notas GPA ###

Comenzamos por generar un histograma para revisar la distribución de las notas GPA, se ha elegido este tipo de gráfico ya que en primera instancia nos interesa revisar estos valores de manera independiente.

```python

sns.histplot(data=df, x="GPA").set_title("Distribución de calificaciones GPA");
```

En el histograma podemos apreciar que los valores se concentran cercanos a la media (1.9) similar a una distribución normal, es posible que con más registros esta pudiese presentar de forma más clara.

### 4. GPA en relación a variables númericas

#### 1. GPA en relación a Age

Se grafica la relación entre las calificaciones y la edad de los estudiantes, en este caso se ha utilizado un boxplot por su facilitada para visualizar distintos estadígrafos.

```python
sns.catplot(data=df, x="Age", y="GPA", hue="Age", kind="box").set(title="GPA vs la edad del estudiante").legend.remove();
```

No se aprecian diferencias significativas entre las distintas edades

#### 2. GPA  en relación a  tiempo de estudio semanal

Se grafica la relación entre las calificaciones de los estudiantes y el tiempo de estudio semanal, además separamos por el campo GradeClass para visualizar con mayor facilidad en que categoría se encuentra cada estudiante.

```python
sns.relplot(data=df, x="StudyTimeWeekly", y="GPA", hue="GradeClass", hue_order = gradeClass).set(title="GPA vs tiempo de estudio");
```

No se observa una correlación tan marcada como se hubiese esperado, podríamos decir que en general los alumnos con la clasificación "A" estudian más y que dentro de los alumnos con clasificación "F" aquellos que más estudian tienen calificaciones ligeramente más altas que el resto de su grupo.

#### 3. GPA  en relación a ausencias

En el caso del gráfico de relación entre calificaciones y ausencias se ha optado por un gráfico de dispersión ya que este nos permite ver de forma rápida si existe una correlación entre las variables, en adición de esto hemos decidido diferenciar los puntos del gráfico mediante el dato GradeClass, el cual nos permite ver rápidamente si un alumnos ha sido aprobado o reprobado (F = Reprobado).

```python
sns.relplot(data=df, x="Absences", y="GPA", hue="GradeClass", hue_order = gradeClass).set(title="GPA vs ausencias");
```

En este caso se observa una fuerte correlación entre la cantidad de ausencias anuales y las calificaciones de los estudiantes, tanto así que podemos realizar afirmaciones como que, dentro de esta muestra nadie con más de diez ausencias ha obtenido una clasificación "A".

### 4. Correlación de GPA con las varibles númericas ###

Como hemos visto una fuerte relación entre las calificaciones de los alumnos y sus ausencias anuales, generamos un mapa de calor para saber qué grado de correlación existe entre estas y visualizar si existe otra correlación que aún no hayamos identificado.

```python
correlacion = df.corr(numeric_only=True)
mask = np.triu(np.ones_like(correlacion, dtype=bool))
cmap = sns.diverging_palette(230, 20, as_cmap=True)
sns.heatmap(correlacion, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5}, annot=True).set(title="Correlación de variables");
```

De acuerdo a lo identificado en el gráfico de dispersión entre las calificaciones y las ausencia, hemos obtenido una alta correlación entre estos campos con un valor de "-0.92" es decir, en general, mientras las ausencias disminuyen las calificaciones de los estudiantes aumentan.
