### 1. Carga de datos ###

Se importan las librerias y se carga dataset limpio

### 3. GPA en relación a variables númericas

#### 1. GPA en relación a Age

```python
sns.catplot(data=df, x="Age", y="GPA", hue="Age", kind="violin");
```

#### 2. GPA  en relación a  tiempo de estudio semanal

```python
sns.relplot(data=df, x="StudyTimeWeekly", y="GPA");
```

```python
sns.relplot(data=df, x="Absences", y="GPA");
```

### 4. Correlación de GPA con las varibles númericas ###

```python
correlacion = df.corr(numeric_only=True)
mask = np.triu(np.ones_like(correlacion, dtype=bool))
cmap = sns.diverging_palette(230, 20, as_cmap=True)
sns.heatmap(correlacion, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5}, annot=True);
```

### 5. Análisis y conclusiones ###

**Análisis GPA en relación a variables categóricas**

En los graficos de las variables categoricas que hemos revisado se puede ver que existen dos factores que afectan de forma leve las calificaciónes de los estudiantes:
* Apoyo parental: En promedio un mayor apoyo parental aumenta las calificiones de los estudiantes
* Existencia de tutor: En promedio si un alumno posee un tutor este tendrá mejor desempeño

Por otra parte contrario a lo que se podría pensar, el nivel de estudio de los padres no parece afectar significativamente la calificaciones de los alumnos, ni tampoco lo hacen los distintos tipo de  actividades extracurriculares

**Análisis GPA en relación a variables númericas**

En cuanto a los graficos de las variables númericas, nos encontramos con que existe una relación mucho menor a la que esperaríamos entre el tiempo de estudio semanal y las calificaciones y por otra parte vemos una fuerte y clara correlación de estas con las usencias.

**Conclusión**

Una clara aplicación de los resultados obtenidos en este análisis es centrar los esfuerzos del establecimiento educacional en disminuir al minimo posible la ausencia de los alumnos ya que esta se encuentra fuertemente relacionada con las calificaciones de estos, tambien se puede solicitar mayor apoyo por parte de los padres e incluso si el establecimiento posee los recursos necesarios asignar un tutor a los alumnos con peores calificaciones, pero siempre teniendo presente que lo más importante es que estos asistan a clases.
